```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import talib
from talib import abstract
import yfinance as yfin
```

```python
df = yfin.download('^TWII', start = '2012-1-1', end = '2024-7-8').round(3)
df.columns = ['open', 'high', 'low', 'close', 'adj_close', 'volume']
tdf = pd.DataFrame()
df
```

```python
df['observed_sell'] = np.nan
df['observed_buy'] = np.nan
df['reverse_sell'] = np.nan
df['reverse_buy'] = np.nan
df['break_buy'] = np.nan
df['break_sell'] = np.nan

for i in range(1, len(df)):
    df['observed_sell'] = df['high'][i - 1] + 0.35 * (df['close'][i - 1] - df['low'][i - 1])
    df['observed_buy'] = df['low'][i - 1] - 0.35 * (df['high'][i - 1] - df['low'][i - 1])
    df['reverse_sell'] = 1.07/2 * (df['high'][i - 1] + df['low'][i - 1]) - 0.07 * df['low'][i - 1]
    df['reverse_buy'] = 1.07/2 * (df['high'][i - 1] + df['low'][i - 1]) - 0.07 * df['high'][i - 1]
    df['break_buy'] = df['observed_sell'][i - 1] + 0.25 * (df['observed_sell'][i - 1] - df['observed_buy'][i - 1])
    df['break_sell'] = df['observed_buy'][i - 1] - 0.25 * (df['observed_sell'][i - 1] - df['observed_buy'][i - 1])
    
df
```
